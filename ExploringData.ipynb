```python
from nltk.corpus import stopwords 
def refineWords(s):
    letters_only = re.sub("[^a-zA-Z]", " ", s) 
    words = letters_only.lower().split()
    stops = set(stopwords.words("english"))
    meaningful_words = [w for w in words if not w in stops]
    #print( " ".join( meaningful_words ))
    return( " ".join( meaningful_words ))
train["text"].fillna(" ",inplace=True)    
train["text"] = train["text"].apply(refineWords)
train["author"].fillna(" ",inplace=True)    
train["author"] = train["author"].apply(refineWords)
train["site_url"].fillna(" ",inplace=True)    
train["site_url"] = train["site_url"].apply(refineWords)
train["title"].fillna(" ",inplace=True)    
train["title"] = train["title"].apply(refineWords)
train["thread_title"].fillna(" ",inplace=True)    
train["thread_title"] = train["thread_title"].apply(refineWords)
train_two = train.copy()
train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000) 
#print(train["title"].head())
#temp  = (vectorizer.fit_transform(train["text"]))
#train["text"] = temp.to_array()
train["text"] = vectorizer.fit_transform(train["text"]).toarray()
train["author"] = vectorizer.fit_transform(train["author"]).toarray()
train["site_url"] = vectorizer.fit_transform(train["site_url"]).toarray()
train["title"] = vectorizer.fit_transform(train["title"]).toarray()
train["thread_title"] = vectorizer.fit_transform(train["thread_title"]).toarray()
train.head()
```

```python
train["isSpam"] = np.sign(train["spam_score"]-0.5)
#print(train_one["isSpam"])
from sklearn.ensemble import RandomForestClassifier
#forest = RandomForestClassifier(n_estimators = 100)
forest = RandomForestClassifier(max_depth = 10, min_samples_split=2, n_estimators = 100, random_state = 1)
features_forest = train[["text", "author", "site_url", "title", "thread_title","domain_rank"]].values
my_forest = forest.fit(features_forest, train["isSpam"])
```

```python
test_features = test[["text", "author", "site_url", "title", "thread_title","domain_rank"]].values
my_prediction = my_forest.predict(test_features)
print(len(my_prediction),len(test["isSpam"]))
```
